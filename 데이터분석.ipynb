```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 설정
months = np.array(['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb'])
sales = np.array([5280000, 5501000, 5469000, 5480000, 5533000, 5554000])
target_sales = np.array([5280000, 5500000, 5729000, 5968000, 6217000, 6476000])

# 매출 추이 그래프
plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o', label='Sales')
plt.plot(months, target_sales, marker='o', label='Target Sales')
plt.title('Sales Trends over the Past 6 Months')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.legend()
plt.show()

# 매출 차이 계산
sales_diff = target_sales - sales

# 매출 차이 그래프
plt.figure(figsize=(10, 6))
plt.bar(months, sales_diff)
plt.title('Difference between Target Sales and Actual Sales')
plt.xlabel('Month')
plt.ylabel('Sales Difference ($)')
plt.show()
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

# 월, 각 변수들의 값을 설정
months_num = np.array([9, 10, 11, 12, 1, 2]).reshape(-1, 1)
ad_costs = np.array([1056000, 950400, 739200, 528000, 316800, 316800]).reshape(-1, 1)
social_costs = np.array([0, 105600, 316800, 528000, 739200, 739200]).reshape(-1, 1)
price_per_ounce = np.array([2.00, 2.00, 2.00, 1.90, 1.90, 1.90]).reshape(-1, 1)
sales = np.array([5280000, 5501000, 5469000, 5480000, 5533000, 5554000]).reshape(-1, 1)

# 데이터 스케일링
scaler = MinMaxScaler()
scaler.fit(np.concatenate([months_num, ad_costs, social_costs, price_per_ounce, sales]))

months_num = scaler.transform(months_num)
ad_costs = scaler.transform(ad_costs)
social_costs = scaler.transform(social_costs)
price_per_ounce = scaler.transform(price_per_ounce)
sales = scaler.transform(sales)

# 데이터 합치기
X = np.concatenate([months_num, ad_costs, social_costs, price_per_ounce], axis=1)

# 다중 선형 회귀 모델 학습
model = LinearRegression()
model.fit(X, sales)

# 미래의 3개월 (3월, 4월, 5월) 매출 예측
future_months = np.array([3, 4, 5]).reshape(-1, 1)
future_ad_costs = np.array([316800, 316800, 316800]).reshape(-1, 1)  # 예시 값
future_social_costs = np.array([739200, 739200, 739200]).reshape(-1, 1)  # 예시 값
future_price_per_ounce = np.array([1.90, 1.90, 1.90]).reshape(-1, 1)  # 예시 값

# 데이터 스케일링
future_months = scaler.transform(future_months)
future_ad_costs = scaler.transform(future_ad_costs)
future_social_costs = scaler.transform(future_social_costs)
future_price_per_ounce = scaler.transform(future_price_per_ounce)

# 데이터 합치기
future_X = np.concatenate([future_months, future_ad_costs, future_social_costs, future_price_per_ounce], axis=1)

# 매출 예측
future_sales = model.predict(future_X)

# 예측된 매출을 원래의 스케일로 되돌리기
future_sales = scaler.inverse_transform(future_sales)

print('Future sales predictions:')
for month, sales in zip(['Mar', 'Apr', 'May'], future_sales):
    print(f'{month}: ${sales[0]:.2f}')
```
